```python
from __future__ import absolute_import, division, print_function, unicode_literals

from agents.DuelingDDQNAgent import DuelingDDQNAgent
from utils import *
import os
import numpy as np
import time
import gym
from gym.wrappers import AtariPreprocessing, FrameStack
%load_ext autoreload
%autoreload 2
```

```python
# env = gym.make('SpaceInvadersNoFrameskip-v4')
# env = gym.make('BreakoutNoFrameskip-v4')
env = gym.make('PongNoFrameskip-v4')
env = AtariPreprocessing(env)
env = FrameStack(env, num_stack=4)

agent = DuelingDDQNAgent(input_shape=(env.observation_space.shape), n_actions=env.action_space.n)
agent.eps = 0.00

load_agent(agent, os.path.join('results', env.game, 'latest_model'))
```

```python
epochs = 100
scores = []
best_score = -np.inf

for i in range(epochs):
    done = False
    observation = env.reset()

    score = 0
    start_time = time.time()
    while not done:
        action = agent.choose_action(observation)
        next_observation, reward, done, info = env.step(action)
        score += reward
        
        if score > best_score:
            best_score = score
        observation = next_observation

    scores.append(score)

    print('Game {} Score {} Best Score {} ε {:.2f}'.format(
        i, score , best_score, agent.eps
    ))
    print ('Time taken for game {:.2f} sec\n'.format(time.time() - start_time))
    
print("------------------------------")
print('Average {}  Best Score {} Worst Score {}'.format(sum(scores) / len(scores), max(scores), min(scores)))
print("------------------------------\n")
```

```python
import matplotlib.pyplot as plt
plt.plot(scores)
plt.xlabel('Epochs')
plt.ylabel('Score')
plt.title('Scores Plot')
plt.show()
```
